## Using lmfit to fit data for one data set

```python
% pylab inline
import pandas as pd
import datahandling
```

#### Reading data from file

```python
time_data, temp_data, torque_data = datahandling.DataFile('03_PHR_Alcamizer.txt').simple_data()
```

Manually 'trim' torque data to only the part of the curve we want to fit ie removing zeros from beginning and dip from end


```python
cut_1 = 15
cut_2 = 1000
def trim(vector):
    return vector[cut_1:cut_2]

time_data = trim(time_data)
temp_data = trim(temp_data)
torque_data = trim(torque_data)
```

#### Using parameter function

```python
from scipy.optimize.lmfit import Parameters, minimize, report_fit
```

```python
p = Parameters()
#          (         Name, Value, Vary,  Min,  Max)    
p.add_many((         'k1',   2.0, True,  0.0, None),
           (         'k2',   5.0, True,  0.0, None),
           (         'k3',  0.35, True,  0.0, None),
           (         'k4',   0.3, True,  0.0, None),
           (         'k5', 0.001, True,  0.0, None),
           (         'k6',   1.0, True,  0.0, None),
           (         'k7',   0.1, True,  0.0, None),
           (         'k8',  0.05, True,  0.0, None),
           (         'k9',   2.0, True,  0.0, None),
           (        'k10',   2.0, True,  0.0, None),
           (        'k11',   1.0, True,  0.0, None),
           (         'UA', 340.0, True,  0.0, None),
           (       'mu_0',  0.03, True,  0.0, None),
           (          'E',6500.0, True,  0.0, None),
           (          'q',   2.5, True,  0.0, None),
           ('prim_stab_0',   0.7, True,  0.0, None))
```

#### Fitting torque data using minimize()

```python
from curve_func_2 import torque_curve, fcn2min_torque, model_curves 
```

```python
result = minimize(fcn2min_torque, p, args=(time_data, torque_data))
```

```python
report_fit(p, show_correl=False)
```

#### Generating and plotting all model outputs

```python
curves = model_curves(p, time_data)
```

```python
HCl, LDH, poly_act, radical, prim_stab, deg_poly, x_link, T, Tm, mu, torque = curves
```

```python
figure()
plot(time_data, torque, label='fitted torque curve')
plot(time_data, torque_data, label='torque data')
legend()

figure()
plot(time_data, Tm, label='fitted temp curve')
plot(time_data, temp_data, label='temp data')
legend()
xlim([0, 31])

figure()
plot(time_data, HCl, '--', label='HCl')
plot(time_data, LDH, label='LDH')
#plot(time_data, poly_act, '-.', label='poly_act')
plot(time_data, radical, ':', label='radical')
plot(time_data, prim_stab, '--', label='prim_stab')
plot(time_data, deg_poly, label='deg_poly')
plot(time_data, x_link, label='x_link')
legend()
xlim([0, 26])
```
