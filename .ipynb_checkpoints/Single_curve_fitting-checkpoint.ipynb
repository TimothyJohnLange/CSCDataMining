```python
data = pd.read_table('03_PHR_Alcamizer.txt', skiprows = 4, sep = ';', usecols = range(6))
```

```python
print data[:1]
```

```python
time_data = data['t [min]'].values
temp_data = data[data.columns[4]].values
torque_data = data['M [Nm]'].values
```

```python
# Temperature viscosity curve parameters
R = 8.314 
Cp = 900
m = 0.5
T_inf = 208
UA = 340
k1 = 2
mu0 = 0.03
E = 6500
T0 = 125
k2 = 5
```

```python
n = 10.
k3 = 0.25
k4 = 0.5
k5 = 0.1
k6 = 0.01
k7 = 1000
HCl_0 = 0.001
LDH_0 = 0.3
poly_act_0 = 2
prim_stab_0 = 1.2
```

```python
from temp_der import *
from kinetics_eq import *
from visc_torque_eq import *
```

```python
time = linspace(0, time_data[len(time_data) - 1], 1000)
```

```python
T = zeros(len(time))
T[0] = T0
Tm = zeros(len(time))
Tm[0] = 200
muV = zeros(len(time))
muV[0] = mu(mu0, E, R, T0, k7, 0)
torqueV = zeros(len(time))
torqueV[0] = torque(k1, muV[0])
```

```python
HCl = zeros(len(time))
HCl[0] = HCl_0
LDH = zeros(len(time))
LDH[0] = LDH_0
poly_act = zeros(len(time))
poly_act[0] = poly_act_0
radical = zeros(len(time))
prim_stab = zeros(len(time))
prim_stab[0] = prim_stab_0
deg_poly = zeros(len(time))
```

```python
for i in range(1,len(time)):
    j = i - 1
    dt = time[i] - time[j]
    
    T[i] = T[j] + dt*der_T(m, Cp, UA, T_inf, T[j])
    Tm[i] = Tm[j] + dt*der_Tm(k2, T[j], Tm[j])
    
    HCl[i] = HCl[j] + dt*der_HCl(n, k3, HCl[j], LDH[j], k4,
                                 poly_act[j])
    LDH[i] = LDH[j] + dt*der_LDH(n, k3, 
                                 HCl[j], LDH[j])
    poly_act[i] = poly_act[j] + dt*der_poly_act(k4, HCl[j], 
                                                poly_act[j])
    radical[i] = radical[j] + dt*der_radical(k4, HCl[j], poly_act[j],
                                             k5, radical[j],
                                             prim_stab[j], k6)
    prim_stab[i] = prim_stab[j] + dt*der_prim_stab(k5, radical[j], 
                                                   prim_stab[j])
    deg_poly[i] = deg_poly[j] + dt*der_deg_poly(k6, radical[j])
    
    muV[i] = mu(mu0, E, R, T[i], k7, deg_poly[i])
    torqueV[i] = torque(k1, muV[i])
    
    
```

```python
plot(time,muV,time,torqueV,time_data,torque_data)
```
